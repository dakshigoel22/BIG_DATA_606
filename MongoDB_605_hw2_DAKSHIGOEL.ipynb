##2.Explain what Redis and MongoDB are.

**○ What type of database is each?**

**MongoDB**

A NoSQL document-oriented database

Stores data as JSON-like documents (BSON format)

Data is organized into collections (not tables)

Supports flexible schema (documents in the same collection can vary)

Type of Database:

- NoSQL

- Document store

**Redis**

An in-memory key–value database. Stores data primarily in RAM. Designed for extremely fast data access. Often used as a caching layer.

Type of Database:
- NoSQL
- Key–value store

In-memory data store

**○ How does each leverage types of storage/memory?**

**MongoDB:**

Stores data primarily on disk. Uses RAM for caching frequently accessed data. Provides built-in persistence and durability. Suitable for large datasets

**Redis:**
Stores data primarily in RAM. Optional disk persistence (snapshots or append-only file). Optimized for speed rather than long-term storage. Limited by available memory.

**○ How are they the same or different?**

They both are same as:  

- Both are NoSQL databases
- Both support flexible data structures
- Both can scale horizontally
- Both are widely used in modern backend systems
- Both can be used together in the same application

They are different as:    

- Data Model: MongoDB → Document-based (JSON-like documents) whereas Redis → Key–value pairs
- Query Capabilities: MongoDB → Rich queries and aggregation pipelines.  Redis → Simple key lookups (limited querying)
- Primary Purpose: MongoDB → Main application database
Redis → Caching, session storage, real-time counters
- Performance: MongoDB → Fast, but disk-based
Redis → Extremely fast, memory-based



**○ When would each typically be used?**
MongoDB is typically used for:
- Storing user data, products, bookings, content
- Applications needing flexible schema
- Large-scale persistent storage
- Primary database in web/mobile apps

Redis is typically used for:

- Caching frequently accessed queries
- Session management
- Rate limiting
- Real-time analytics

**○ What are the tradeoffs for each?**

MongoDB Tradeoffs:

- Slower than in-memory databases
- Requires indexing for optimal performance
- Can become heavy under large read loads without caching

Redis Tradeoffs:

- Limited by available RAM
- Not ideal for long-term storage
- Cache invalidation can be complex
- Less powerful querying compared to MongoDB

##6.Cache with Redis
○ Imagine now that your Mongo instance has become overly burdened and you need a caching layer for faster querying. Explain why a cache is useful in general and how Redis could be used in conjunction with Mongo to help speed things up.

Why Cache is useful :

A cache is a fast, temporary storage layer that stores frequently accessed data in memory instead of repeatedly querying the primary database.

When MongoDB becomes heavily loaded - Many users may request the same data (e.g., popular movies)

- Complex aggregation queries may be expensive
- Disk I/O becomes a bottleneck

Instead of repeatedly querying MongoDB, we store frequently requested results in a cache (like Redis). Since cache is in-memory, responses are much faster (milliseconds).

Why use Redis for Cache:

- Redis is very fast(in-memory)
- It is key-value based
- Supports TTL - AUTOMATIC EXPIRATION
- It is lightweight and simple to integrate

How Redis+MongoDB work togethor in conjugation:



```
Client->Application->Redis(Cache check)->MongoDB(if cache miss)
```

**■ What are the tradeoffs in using this caching layer?**

1. Cache Invalidation (Biggest Problem) : If data changes in MongoDB:
- Cached data may become stale
- Users may see outdated results

So one must decide when to invalidate? or When to refresh?

2. Extra Complexity : Have to manage all -  MongoDB, Redis, Synchronization logic and Expiration policies.

3. Memory Cost - Redis stores data in RAM. RAM is expensive and limited. Thus we cannot cache everything.


■ What might you need to think about in terms of how you model the data in your cache or otherwise set it up?

1. Caching should be done for Frequently accessed movies, Popular search results, Expensive aggregation queries. We should not cache, Rarely accessed data, Highly volatile data (changes every second)

2. Cache Key Design - Keys should be meaningful and structured.

3. TTL Strategy : Use expiration time , 5 minutes for movie details, 1 hour for top-rated list. Very short TTL for live updates. Too long, stale data . Too short, constant DB hits.

4. Data Modeling in Cache. Redis is key-value, so

Option A — Store whole document as JSON

Option B — Store specific fields

Option C — Store precomputed aggregates

Example - Instead of running aggregation every time.

````python
import redis
r = redis.Redis(host='localhost', port=6379, db=0)
````

````python
import json
#Caching query results : Total movie count,oldest movie, Top 5 movies after 2000

total_movies = movies.count_documents({})
r.set("total_movies", total_movies)

#Cache-2 :
oldest_movie = list(
    movies.find({}, {"title": 1, "year": 1, "_id": 0})
          .sort("year", 1)
          .limit(1)
)

print("Caching oldest_movies")
r.set("oldest_movie", json.dumps(oldest_movie))

#Cache-3:
print("Caching recent_movies")
recent_movies = list(
    movies.find({"year": {"$gt": 2000}},
                {"title": 1, "year": 1, "_id": 0}).limit(5)
)

r.set("recent_movies_2000", json.dumps(recent_movies))
````

````python
# Retrieve Cached Query Result

cached_oldest = r.get("oldest_movie")

if cached_oldest:
    result = json.loads(cached_oldest)
    print("Retrieved from cache:")
    print(result)
else:
    print("Cache miss")

````

### Issues with Caching

Data Science Considerations When Using Cached Data

When performing data analysis on top of a Redis caching layer instead of directly querying MongoDB, several important concerns arise.

1. Stale or Outdated Data

Cached data may not reflect the most recent updates made in the primary database. Since Redis typically stores data temporarily with a time-to-live (TTL), there can be a delay between updates in MongoDB and updates in the cache. If the cache is not invalidated or refreshed appropriately, analysts may unknowingly work with outdated information. This can lead to inaccurate insights, incorrect reporting, and flawed decision-making.


2. Incomplete or Biased Dataset

A cache usually contains only frequently accessed queries or precomputed results, not the entire dataset. As a result, performing analysis on cached data may introduce bias because it represents only a subset of the data. Rare records or less frequently accessed entries may be excluded from the cache, leading to incomplete analysis and potentially skewed conclusions.

3. Reduced Granularity and Flexibility

Cached data is often stored in aggregated or simplified form to improve performance. While this is beneficial for fast retrieval, it limits the ability to perform deeper or custom analysis. Analysts may not have access to raw-level data necessary for advanced statistical modeling, exploratory data analysis, or reproducibility. This reduces analytical flexibility and may require returning to the primary database for more detailed queries.

4. Lack of Data Lineage and Metadata

Cached results may not include information about when they were generated, what query produced them, or what transformations were applied. This lack of metadata makes it difficult to ensure reproducibility, validate findings, or audit analytical processes. For data science workflows, transparency and traceability are critical.
